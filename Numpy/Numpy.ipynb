### Question 4

![image.png](https://raw.githubusercontent.com/rulercosta/Python-Essentials/main/Numpy/Q4.png)

$ P / \left( A \sigma \epsilon \right) = T^4 $

$ P / \left( A \sigma \epsilon {T_0}^4 \right) = \left( \frac{1}{1 + e^{-kt}} \right) ^4 $

```python
kt = np.linspace(0,3,100)
T = 1/(1 + np.exp(-kt))
P = T**4
dkt = kt[1] - kt[0]
E = np.cumsum(P) * dkt
```

```python
plt.plot(kt,E)
plt.xlabel('$kt$')
plt.ylabel(r'$ \left( \frac{k}{A \sigma \epsilon {T_0}^4} \right) E(kt) $')
plt.show()
```

```python
E = np.array([Ex,Ey,Ez])
E = np.swapaxes(E,0,-1)
```

```python
z = np.array([0,1,1])
B = np.cross(z,E)
```

```python
E = np.swapaxes(E,0,-1)
B = np.swapaxes(B,0,-1)
```

```python
Bx, By, Bz = B
```
