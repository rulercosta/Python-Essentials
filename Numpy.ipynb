Integrals and Derivatives

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y,x)
int_y = np.cumsum(y) *(x[1]-x[0])
plt.plot(x,y)
plt.plot(x,dydx)
plt.plot(x,int_y)
```
